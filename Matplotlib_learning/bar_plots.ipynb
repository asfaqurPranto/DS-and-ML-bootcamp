```python
# in bar graph x axis contain namme and y axis contain value for example we will 
student={'Pranto':50,'Torikul': 60,'Omi': 70}  #creating a dictionalry which contains the student name and their number

Names=list(student.keys()) #extract key from dictionary
Marks=list(student.values())


```

```python
from matplotlib import pyplot as plt


plt.bar(Names,Marks)
# title ,x-axis,y-axis,grid is same as line plot``
plt.show()
```
