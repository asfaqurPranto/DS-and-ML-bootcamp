```python
#line plot we need to find value for x axis and some vaue from y axis, in this case 2 list will contain those value 

import numpy as np
from matplotlib import pyplot as plt  # sublibrary importing

x=np.arange(1,11)  # will create array from 1 to 10
y=3*x  #linear operation on array
z=2*x
plt.plot(x,y)   
plt.show()

```

```python
plt.plot(x,y)
plt.plot(x,z)   
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
# we can also modify the colour of the line plt.plot(x,y,colour='g',;inestyle=':',linewidth=2)
plt.show()

 
```

```python
# adding sub-plots
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,z)

plt.show()
```
